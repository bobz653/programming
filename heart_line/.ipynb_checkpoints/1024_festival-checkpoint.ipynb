```python
import numpy as np
import matplotlib.pyplot as plt
T = np.linspace(0 , 2 * np.pi, 1024)
plt.axes(polar = True)
plt.plot(T, 1. - np.sin(T),color="r")
plt.show()

#原文：https://blog.csdn.net/wireless_com/article/details/69817498 
```

```python
import numpy as np
import matplotlib.pyplot as plt
t = np.linspace(0, 2*np.pi, 1024)
Y,X = (2*np.cos(t) - np.cos(2*t))/2,2*np.sin(t) - np.sin(2*t)
plt.plot(X, Y, c = 'r')
plt.show()
```
